# **Problem Statement**


The problem in global terrorism is the persistent threat posed by acts of terrorism across different regions of the world. Terrorism is characterized by the intentional use of violence, intimidation, and fear to achieve political, ideological, or religious objectives. It poses a significant challenge to national security, public safety, and social cohesion.

The objective of the Exploratory Data Analysis(EDA) is to understand the terrorist events around the world. Terrorism continues to be a pressing global issue with far-reaching consequences for human lives, security, and social stability. However, the lack of a comprehensive and standardized dataset on global terrorism incidents poses significant challenges in understanding the patterns, trends, and underlying factors associated with this phenomenon. To address these challenges, there is a critical need for a project focused on creating a Global Terrorism Dataset that collects, cleans, standardizes, and analyzes data on terrorism incidents worldwide. By establishing a reliable foundation for research and analysis, this dataset would enable a deeper understanding of terrorism, facilitate evidence-based policymaking, and support the development of effective counterterrorism strategies to mitigate the threat posed by terrorism globally.

As we can see, 'Wounded' and 'Killed' Attributes have lots of Null values, need to fill null values

````python
terrorism_df['Wounded'] = terrorism_df['Wounded'].fillna(0).astype(int)
terrorism_df['Killed'] = terrorism_df['Killed'].fillna(0).astype(int)
terrorism_df['Latitude'] = terrorism_df['Latitude'].fillna(0).astype(int)
terrorism_df['Longitude'] = terrorism_df['Longitude'].fillna(0).astype(int)
````
